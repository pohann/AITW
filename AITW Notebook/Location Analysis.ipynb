```python
import pandas as pd


import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./F2F Donors1023(Excel editing).csv',encoding = 'cp950',dtype={12:'int8'})
df.head()
```

```python
#deleting donors that have nan values for important variables
df_temp = df[pd.isna(df['First Debit Date'])==False].reset_index()
df_temp = df_temp.drop(['index'],axis=1)
#df_temp = df_temp[pd.isna(df_temp['Sign Up Date'])==False].reset_index()
#df_temp = df_temp.drop(['index'],axis=1)
df_temp = df_temp[pd.isna(df_temp['Meeting Point'])==False].reset_index()
df_temp = df_temp.drop(['index'],axis=1)
```

```python
df_temp.head()
```

```python
#add labels to indicate wether a donor has successful first debit

from datetime import date
import numpy as np


import warnings
warnings.filterwarnings('ignore')

df_temp['Successful First Debit'] = np.zeros(df_temp.shape[0])


for i in range(df_temp.shape[0]):
    FDD = df_temp['First Debit Date'][i].split('/')
    FDD = date(int(FDD[0]),int(FDD[1]),int(FDD[2]))
    if pd.isna(df_temp['First Failed Debit'][i]) == False:
        FFD = df_temp['First Failed Debit'][i].split('/')
        FFD = date(int(FFD[0]),int(FFD[1]),int(FFD[2]))
        dif = FFD - FDD
        if dif.days > 10:
            df_temp['Successful First Debit'][i] = 1
    else:
        df_temp['Successful First Debit'][i] = 1
df_temp['Successful First Debit'].sum()
```

```python
df_temp['Meeting Point'].unique()
#南?角=南勢角
#頭前?=頭前庄
#行天?=行天宮
```

```python
for place in df_temp['Meeting Point'].unique():
    print(str(place)+' '+str(df_temp[df_temp['Meeting Point'] == place][df_temp['Payment Method']=='DEBIT']['Successful First Debit'].mean()))
```

```python
df_temp.to_csv('./F2F Donors(new).csv',encoding='cp950',index=False)
```

# Get nearest MRT station

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
df_temp = pd.read_csv('./F2F Donors1023.csv',encoding='cp950')
df_temp.head()
```

```python
df_temp['MRT station'] = np.zeros(df_temp.shape[0])
```

```python
from selenium import webdriver
#driver = webdriver.Firefox()
driver = webdriver.Chrome('./chromedriver.exe')
driver.implicitly_wait(5)
```

```python
from time import sleep
#convert meeting point to nearest MRT station

error_dict = {'行天?':'行天宮','南?角':'南勢角','頭前?':'頭前庄'}
for i in range(20546,df_temp.shape[0]):
    if df_temp['MRT station'][i] == 0 and pd.isnull(df_temp['Meeting Point'][i]) == False:
        driver.get('https://www.google.com/')
        search = driver.find_elements_by_xpath('//*[@title="Google 搜尋"]')

        if df_temp['Meeting Point'][i] in error_dict.keys():
            location = error_dict[df_temp['Meeting Point'][i]]
        else:
            location = str(df_temp['Meeting Point'][i])
        search[0].send_keys(location+' 附近捷運站')
        sleep(0.5)
        search_button = driver.find_elements_by_xpath('//*[@value="Google 搜尋"]')
        search_button[1].click()
        sleep(1)
        MRT = driver.find_elements_by_xpath('//*[@class="dbg0pd"]')
        try:
            df_temp['MRT station'][i] = MRT[0].text
        except IndexError:
            continue
```

```python
for i in range(df_temp.shape[0]):
    if df_temp['MRT station'][i] == '沙鹿車站':
        df_temp['MRT station'][i] = '林口站'
```
