```python
supporter_df = pd.read_csv('./20191015_Supporter_All.csv',encoding = 'cp950')
supporter_df.head()
```

```python
for i in range(len(supporter_df['群組'])):
    if supporter_df['群組'].isnull()[i] == False:
        for tag in tag_list[2][0:6]:
            if tag.split(',')[0] in supporter_df['群組'][i].split(',') and tag.split(',')[1] in supporter_df['群組'][i].split(','):
                supporter_df['Actions'][i] -= 1 
```

```python
verify = 0
for i in range(len(supporter_df['群組'])):
    if supporter_df['群組'].isnull()[i] == False:
            if tag_list[2][3].split(',')[0] in supporter_df['群組'][i].split(',') and tag_list[2][3].split(',')[1] in supporter_df['群組'][i].split(','):
                verify += 1
print(verify)
```
